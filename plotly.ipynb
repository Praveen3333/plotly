````python


### Q1: Load the "titanic" dataset using the `load_dataset` function of seaborn. Use Plotly Express to plot a scatter plot for the age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot scatter plot
fig = px.scatter(titanic, x='age', y='fare', title='Scatter Plot of Age vs Fare')
fig.show()
```

### Q2: Using the tips dataset in the Plotly library, plot a box plot using Plotly Express.

```python
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Plot box plot
fig = px.box(tips, x='day', y='total_bill', title='Box Plot of Total Bill by Day')
fig.show()
```

### Q3: Using the tips dataset in the Plotly library, plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
# Plot histogram
fig = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker", title='Histogram of Total Bill by Sex')
fig.show()
```

### Q4: Using the iris dataset in the Plotly library, plot a scatter matrix plot, using the "species" column for the color parameter. Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.

```python
# Load the iris dataset
iris = px.data.iris()

# Plot scatter matrix
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species", title='Scatter Matrix of Iris Dataset')
fig.show()
```

### Q5: What is a Distplot? Using Plotly Express, plot a distplot.

**Distplot**: A distplot is a combination of a histogram and a kernel density estimate (KDE) plot, used to visualize the distribution of a dataset. However, Plotly Express does not have a direct function named `distplot`. We can create a similar plot using a combination of histograms and KDE plots.

```python
import plotly.figure_factory as ff

# Generate random data
import numpy as np
data = np.random.randn(1000)

# Plot distplot
fig = ff.create_distplot([data], group_labels=['distplot'], show_hist=True, show_curve=True)
fig.show()
```
````
